```python
files =[]
files = glob.glob(r'..\Data\*.csv')
# print(*files,sep='\n') # Printing name of all the files
print("Total Files Read : ",len(files)) # To check if all 88 datasets are read 
```

```python
list_df =[]
c = ['avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23']
for file in files:
    df = pd.read_csv(file,skiprows=4,usecols=c)
    name = file.split('\\')[-1].split('.')[0] # Name of dataset
    st = df.describe().drop('count').transpose()
    instance = pd.concat([st.loc['avg_rss12'],st.loc['var_rss12'],st.loc['avg_rss13'],st.loc['var_rss13'],st.loc['avg_rss23'],
                          st.loc['var_rss23']],axis=0, sort=False)
    instance = instance.to_frame().transpose()
    instance.columns = ['Min_1','Max_1','Mean_1','Median_1','Std_1','FirstQuartile_1', 'ThirdQuartile_1',
                        'Min_2','Max_2','Mean_2','Median_2','Std_2','FirstQuartile_2', 'ThirdQuartile_2',
                        'Min_3','Max_3','Mean_3','Median_3','Std_3','FirstQuartile_3', 'ThirdQuartile_3',
                        'Min_4','Max_4','Mean_4','Median_4','Std_4','FirstQuartile_4', 'ThirdQuartile_4',
                        'Min_5','Max_5','Mean_5','Median_5','Std_5','FirstQuartile_5', 'ThirdQuartile_5',
                        'Min_6','Max_6','Mean_6','Median_6','Std_6','FirstQuartile_6', 'ThirdQuartile_6']
    instance['File Name'] = name # Column representing name of dataset
    list_df.append(instance)
    final_df = pd.concat(list_df)

# Creating instance number as index

final_df['instance'] = np.arange(start = 1,stop = len(final_df)+1)
final_df.set_index('instance',inplace=True)

final_df.fillna(value=0,inplace=True) # Imputed NaN's with 0.
final_df
```

```python
std_features = final_df.std()
std = pd.DataFrame(std_features,columns=['Standard Deviation'])
print("Standard Deviation for all time domain features : ")
std
```

```python
# Checking if computation of all the metrics is correct by cross checking with one dataset.
col = ['avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23']
d = pd.read_csv(r'..\Data\dataset9w.csv',skiprows=4,usecols=col)

stat = d.describe().drop('count')
mean1 = stat['avg_rss12'][0] # MEAN
stdev1 = stat['avg_rss12'][1] # STD
mn1 = stat['avg_rss12'][2] # MIN
firqt1 = stat['avg_rss12'][3] # 25%
mdn1 = stat['avg_rss12'][4] # MEDIAN(50%)
thrqt1 = stat['avg_rss12'][5] # 75%
mx1 = stat['avg_rss12'][6] # MAX

col_names = ["Instance","Mean","Standard Deviation", "Min","First Quartile","Median","Third Quartile","Max"]

table_data = [[0,mean1,stdev1,mn1,firqt1,mdn1,thrqt1,mx1]]
        
print(tabulate(table_data, headers=col_names, tablefmt="grid",floatfmt=".7f"))
```

```python
final_df2 = final_df.drop('File Name',axis=1)
final_df2.reset_index(inplace=True)
final_df2 = final_df2.drop('instance',axis=1)

res = bootstrap((final_df2,),np.std,confidence_level=0.9,random_state=0,method='basic')
ci = res.confidence_interval
ci_df = pd.DataFrame()
ci_df["Name"] = ['Min_1','Max_1','Mean_1','Median_1','Std_1','FirstQuartile_1', 'ThirdQuartile_1',
                        'Min_2','Max_2','Mean_2','Median_2','Std_2','FirstQuartile_2', 'ThirdQuartile_2',
                        'Min_3','Max_3','Mean_3','Median_3','Std_3','FirstQuartile_3', 'ThirdQuartile_3',
                        'Min_4','Max_4','Mean_4','Median_4','Std_4','FirstQuartile_4', 'ThirdQuartile_4',
                        'Min_5','Max_5','Mean_5','Median_5','Std_5','FirstQuartile_5', 'ThirdQuartile_5',
                        'Min_6','Max_6','Mean_6','Median_6','Std_6','FirstQuartile_6', 'ThirdQuartile_6']

ci_df["Lower Bound"] = ci[0]
ci_df["Upper Bound"] =ci[1]

print("Confidence Interval")
ci_df
```
